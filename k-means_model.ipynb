```python
#Histograms
import matplotlib.pyplot as plt
df_without_id = df.drop('Segment ID', axis=1)
df_without_id.hist(bins=30, figsize=(10,10))
plt.show()
```

```python
#Outlier removal using IQR
Q1 = df_without_id.quantile(0.25)
Q3 = df_without_id.quantile(0.75)
IQR = Q3 - Q1
df_without_id_outliers_removed = df_without_id[~((df_without_id < (Q1 - 1.5 * IQR)) | (df_without_id > (Q3 + 1.5 * IQR))).any(axis=1)]
df_without_id_outliers_removed.boxplot(figsize=(12, 8))
plt.show()
```

```python
#Outlier removal using Z-score
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(df_without_id))
df_without_id_outliers_removed_z = df_without_id[(z < 3).all(axis=1)]
df_without_id_outliers_removed_z.boxplot(figsize=(12, 8))
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

#Loading dataset using Pandas
df = pd.read_csv('dataset_working.csv')

#Igonring the Segment ID column
columns_check = [col for col in df.columns if (col != 'Segment ID')]

Q1 = df[columns_check].quantile(0.25)
Q3 = df[columns_check].quantile(0.75)
IQR = Q3 - Q1

#Removing outliers
df_without_outliers = df[~((df[columns_check] < Q1 - 1.5 * IQR) | (df[columns_check] > Q3 + 1.5 * IQR)).any(axis=1)]

#Square root transformation
df_without_outliers['p4'] = np.sqrt(df_without_outliers['p4'])
df_without_outliers['p5'] = np.sqrt(df_without_outliers['p5'])
df_without_outliers['p6'] = np.sqrt(df_without_outliers['p6'])
df_without_outliers['p8'] = np.sqrt(df_without_outliers['p8'])

#Log transformation for p9
df_without_outliers['p9'] = np.log1p(df_without_outliers['p9'])  

#Standardization for features
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_without_outliers[['p1', 'p2', 'p3', 'p4', 'p5', 'p6', 'p7', 'p8', 'p9']])

#Applying PCA with 95% variance
pca = PCA(n_components=0.95)
principalComponents = pca.fit_transform(df_scaled)
print("Number of components: ", principalComponents.shape[1])
```

```python
#Applying KMeans
kmeans = KMeans(n_clusters=2, init='k-means++', max_iter=400, n_init=10, random_state=42)
kmeans_output = kmeans.fit_predict(principalComponents)
#print(kmeans_output)

#Reset index number of updated dataframe
df_without_outliers.reset_index(drop=True, inplace=True)

#Labels for clusters
custom_labels = {0: 'B', 1: 'A'}

#Adding cluster column to dataframe
df_without_outliers['cluster'] = pd.Series(kmeans_output).map(custom_labels)

print(df_without_outliers.count())
print(df_without_outliers)
```

```python
#putting cluster1 and cluster2 in separate dataframes
cluster1 = df_without_outliers[df_without_outliers['cluster'] == 'A']
cluster2 = df_without_outliers[df_without_outliers['cluster'] == 'B']

#getting count of each cluster
count_cluster1 = cluster1['cluster'].count()
count_cluster2 = cluster2['cluster'].count()

print("Count of Cluster 1: ", count_cluster1)
print("COunt of Cluster 2: ", count_cluster2)

#loading testing dataset
df_test = pd.read_csv('dataset_testing.csv')

#create a dataframe with segment ID of testing dataset that are only in cluster1 and cluster2
df_test_cluster_1_and_2 = df_test[df_test['Segment ID'].isin(df_without_outliers['Segment ID'])]
#print(df_test_cluster_1_and_2)

#get count of each label in testing dataset
count_Test_Awake = df_test_cluster_1_and_2['label'].value_counts()['Awake']
count_Test_Deep_anaesthetic = df_test_cluster_1_and_2['label'].value_counts()['Deep anaesthetic ']

print("Count of TestData Awake: ", count_Test_Awake)
print("Count of TestData Deep anaesthetic: ", count_Test_Deep_anaesthetic)
print("Total Count", len(df_test_cluster_1_and_2))

count_fault_cl1 = 0
count_fault_cl2 = 0
count1 = 0
count2 = 0

#dataframe for cluster1 and cluster2
df_clusters = pd.DataFrame(columns=['Segment ID', 'cluster', 'label'])

#calculate error rate of cluster1
for i in df_test_cluster_1_and_2['Segment ID']:
    if i in cluster1['Segment ID'].values:  # Use .values to compare with the actual array of values
        print("Cluster 1 Seg ID - ", cluster1.loc[cluster1['Segment ID'] == i, 'Segment ID'].values[0], " Custer Label - ", cluster1.loc[cluster1['Segment ID'] == i, 'cluster'].values[0], "Test Dataset Seg ID ", df_test.loc[df_test['Segment ID'] == i, 'Segment ID'].values[0], " Test Dataset Label - ", df_test.loc[df_test['Segment ID'] == i, 'label'].values[0])
        #addig to dataframe
        df_clusters = pd.concat([df_clusters, pd.DataFrame({'Segment ID': [i], 'cluster': [cluster1.loc[cluster1['Segment ID'] == i, 'cluster'].values[0]], 'label': [df_test.loc[df_test['Segment ID'] == i, 'label'].values[0]]})], ignore_index=True)
        if df_test.loc[df_test['Segment ID'] == i, 'label'].values[0] != "Awake":
            count_fault_cl1 += 1
    

# #calculate error rate of cluster2
for i in df_test_cluster_1_and_2['Segment ID']:
    if i in cluster2['Segment ID'].values:  # Use .values to compare with the actual array of values
        print("Cluster 2 Seg ID - ", cluster2.loc[cluster2['Segment ID'] == i, 'Segment ID'].values[0], " Custer Label - ", cluster2.loc[cluster2['Segment ID'] == i, 'cluster'].values[0], "Test Dataset Seg ID ", df_test.loc[df_test['Segment ID'] == i, 'Segment ID'].values[0], " Test Dataset Label - ", df_test.loc[df_test['Segment ID'] == i, 'label'].values[0])
        #addig to dataframe
        df_clusters = pd.concat([df_clusters, pd.DataFrame({'Segment ID': [i], 'cluster': [cluster2.loc[cluster2['Segment ID'] == i, 'cluster'].values[0]], 'label': [df_test.loc[df_test['Segment ID'] == i, 'label'].values[0]]})], ignore_index=True)
        if df_test.loc[df_test['Segment ID'] == i, 'label'].values[0] != "Deep anaesthetic ":
            count_fault_cl2 += 1

print("Count of fault in cluster1: ", count_fault_cl1)
print("Count of fault in cluster2: ", count_fault_cl2)

#calculate error rate for each cluster
error_rate_cluster1 = count_fault_cl1/count_cluster1 * 100
error_rate_cluster2 = count_fault_cl2/count_cluster2 * 100

#calculate error rate for each label in testing dataset
error_rate_Test_Awake = count_fault_cl1/count_Test_Awake * 100
error_rate_Test_Deep_anaesthetic = count_fault_cl2/count_Test_Deep_anaesthetic * 100

print ("Error rate of Cluster 1: ", error_rate_cluster1)
print("Error rate of Cluster 2: ", error_rate_cluster2)

print("Error rate of Test Dataset label Awake: ", error_rate_Test_Awake)
print("Error rate of Test Dataset label Deep Anaesthetic : ", error_rate_Test_Deep_anaesthetic)

print(df_clusters)

#saving the dataframe to xlsx
df_clusters.to_excel("k-means.xlsx", index=False)
```
